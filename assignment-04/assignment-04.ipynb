```python
import matplotlib.pyplot as plt
import pandas as pd

# read in file with semicolon delimiter
df = pd.read_csv('data/store.csv', encoding='latin-1')

print(df.dtypes)
print('\nRows=' + str(len(df.index)) + '\n')
# select only cardinal data
df_card = df[['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping.Cost']]
# sort ascending
df_card = df_card.reindex(sorted(df_card.columns), axis=1)

# calc correlation matrix
cor_matrix = df_card.corr()

# plot the matrix
f = plt.figure(figsize=(20, 20))
plt.matshow(cor_matrix, fignum=f.number)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=25)
plt.xticks(range(df_card.select_dtypes(['number']).shape[1]), df_card.select_dtypes(['number']).columns, fontsize=25,
           rotation=90)
plt.yticks(range(df_card.select_dtypes(['number']).shape[1]), df_card.select_dtypes(['number']).columns, fontsize=25)
plt.show()
f.savefig('exports/1_2_matrix_figure.png')

cor_matrix
temp_df = df.groupby(['Order.ID'])
print('\nRows=' + str(len(temp_df)) + '\n')
df.head()
```
